```python
import yfinance as yf

test_data = yf.download("BTC-USD", period="5d", interval="1h")
test_data.head()
```

```python
test_data.columns = test_data.columns.get_level_values(0)
test_data.head()
```

```python
import pandas_ta as ta

test_data["SMA_20"] = ta.sma(test_data.ta.sma(close=test_data["Close"], length=20))

test_data.tail()
```

```python
test_data["RSI_14"] = test_data.ta.rsi(close=test_data["Close"], length=14)
test_data.tail()
```
